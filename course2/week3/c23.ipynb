This week we'll find the Pearson correlation coefficient for two numeric variables. Since the Outlook On Life data set used for the previous weeks and courses lacks two quantitative variables to compare, we'll use the Mars Crater data set instead.

```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sb
import scipy as sp
import matplotlib.pyplot as plt
```

```python
#generate dataframe

df = pd.read_csv('mymarsnotebook.csv', low_memory=False)
```

```python
#coax the data into a numerical format

df['DIAM_CIRCLE_IMAGE'] = df['DIAM_CIRCLE_IMAGE'].apply(pd.to_numeric, errors='coerce' )
df['DEPTH_RIMFLOOR_TOPOG'] = df['DEPTH_RIMFLOOR_TOPOG'].apply(pd.to_numeric, errors='coerce' )
```

We create a scatterplot to look for patterns of association in the data.

```python
#scatterplot:  Q->Q
scat2 = sb.regplot(x='DEPTH_RIMFLOOR_TOPOG', y='DIAM_CIRCLE_IMAGE',  fit_reg=False, data=df)
plt.xlabel('DEPTH_RIMFLOOR_TOPOG')
plt.ylabel('DIAM_CIRCLE_IMAGE')
plt.title('Scatterplot for the Association Between Crater Rimfloor Depth and Crater Circle Image Diameter')
```

We notice that there seems to be at least a weakly linear association between the Crater Rimfloor Depth and Crater Image Circle Diameter. Finally, we find the Pearson correlation coefficient.

```python
print ('Association between Crater Depth and Crater Diameter')
print (sp.stats.pearsonr(df['DEPTH_RIMFLOOR_TOPOG'], df['DIAM_CIRCLE_IMAGE']))
```

### Summary:
---
The Pearson correlation coefficient of ~ 0.59 corroborates our observation from the scatterplot that there is indeed a linear correlation between the Mars Crater Rimfloor Depth and Crater Image Circle Diameter.
